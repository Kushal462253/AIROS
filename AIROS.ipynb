## 1. 🚀 Introduction

Scientific research is moving faster than ever, but the tools researchers use to discover, read, synthesize, and compile literature remain fragmented. **AIROS** (AI Research Operating System) is designed to solve this by creating a centralized, persistent workspace powered by multi-agent reasoning.

Instead of functioning as a simple chat window, AIROS maps scientific inquiry onto structured, persistent workspaces. Distinct agents analyze research intent, retrieve papers across academic databases (arXiv and Semantic Scholar), parse and index PDF manuscripts, build relationship knowledge graphs, and support natural-language copilot navigation grounded on actual papers.

## 2. ⚠️ Problem Statement

Modern researchers face an information overload. When starting a new project, they must:

1. **Search** academic indexes like arXiv and Semantic Scholar.
2. **Download** dozens of PDFs and manually upload them into document viewers.
3. **Analyze** papers to extract claims, evidence, and experimental metrics.
4. **Organize** findings into folders, notes, or spreadsheets.
5. **Synthesize** links between different papers and research questions.

This process is highly disjointed, leading to cognitive fatigue, loss of context across sessions, and difficulty tracking citations.

## 3. 🧩 Why Current Research Workflows are Fragmented

Traditional workflows rely on single-purpose utility tools:

* **Generic Chatbots:** Suffer from hallucination of scientific citations because they are not grounded in real-time academic databases or user-uploaded documents.
* **Disconnected Tools:** Searching happens in browser tabs, paper storage in Zotero, note-taking in Notion, and analysis in ChatGPT. There is no shared semantic memory.
* **Manual Evidence Tracking:** Finding which paper made which claim, and validating its statistical confidence, requires manually searching through PDF text again and again.

## 4. 💡 The AIROS Solution

AIROS centralizes these workflows into a single, cohesive **AI Research Operating System**:

* **State Persistence:** Workspaces are saved and persist across sessions, ensuring context is never lost.
* **Grounded RAG Pipeline:** Search results and PDF files are ingested and vectorized directly, ensuring the AI Copilot answers are grounded with explicit sources.
* **Simulated Multi-Agent Collaboration:** A sequenced execution pipeline decomposes research objectives and drives retrieval, claim extraction, and visual networking.

## 6. 🤖 Multi-Agent Workflow

AIROS operates a sequence of specialized AI agents:

1. **Planner Agent:** Formulates formal objectives, research questions, and runtime estimates from the user's project intent.
2. **Retrieval Agent:** Fetches matching papers from arXiv, Semantic Scholar, and Web Index.
3. **Evidence Agent:** Extracts claims, metrics, and quotes, rating their confidence.
4. **Knowledge Graph Agent:** Synthesizes semantic nodes (Objective, Paper, Concept, Claim) and draws relationships.
5. **Memory Agent:** Persists session histories and indexing parameters.
6. **Copilot Agent:** Answers queries, citations, and navigation instructions.

* **Literature Search Aggregator:** Queries arXiv and Semantic Scholar APIs simultaneously, calculating local relevance scores.
* **Vector PDF Ingestion:** Extracted text is chunked and stored in ChromaDB (or local TF-IDF fallback) for Retrieval-Augmented Generation (RAG).

* **Evidence Ledger:** Extracts key scientific claims with supporting verbatim quotes, confidence ratings, and paper mappings.
* **Grounded Copilot:** Reasoning engine that evaluates query context and presents confidence scores and source citations.

### A. Custom Vector Store TF-IDF Fallback Engine

This lightweight pure-python fallback implements term frequency calculations and cosine similarity for document matches when external ChromaDB dependencies are not installed.

````python
import math
import re
from typing import List, Dict, Any

class FallbackVectorStoreDemo:
    """A lightweight representation of the AIROS Fallback Vector Store."""
    def __init__(self, vocab: List[str]):
        self.vocab = {word: idx for idx, word in enumerate(sorted(vocab))}
        # Math IDF initialization
        self.idf = [1.5] * len(self.vocab)

    def _tokenize(self, text: str) -> List[str]:
        return re.findall(r"\b[a-zA-Z0-9_]+\b", text.lower())

    def _vectorize(self, tokens: List[str]) -> List[float]:
        vector = [0.0] * len(self.vocab)
        tfs = {}
        for token in tokens:
            if token in self.vocab:
                tfs[token] = tfs.get(token, 0) + 1
        for term, count in tfs.items():
            t_idx = self.vocab[term]
            vector[t_idx] = count * self.idf[t_idx]
        sq_sum = sum(x*x for x in vector)
        if sq_sum > 0:
            norm = math.sqrt(sq_sum)
            vector = [x / norm for x in vector]
        return vector

    def calculate_similarity(self, doc_vec: List[float], query_vec: List[float]) -> float:
        # Cosine similarity for L2 normalized vectors is the dot product
        return sum(doc_vec[i] * query_vec[i] for i in range(len(self.vocab)))

# Instantiating and executing a simple validation run
vocab_sample = ["quantum", "qubit", "cryogenic", "temperature", "computing", "error", "correction"]
store = FallbackVectorStoreDemo(vocab_sample)
v1 = store._vectorize(["quantum", "error", "correction"])
v2 = store._vectorize(["cryogenic", "quantum", "computing"])
similarity = store.calculate_similarity(v1, v2)
print(f"Validated Cosine Similarity Score: {similarity:.4f}")
````
